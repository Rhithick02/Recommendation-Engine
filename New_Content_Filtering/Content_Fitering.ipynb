```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('final2.csv')
df.dropna(how = 'any')
df = df.drop_duplicates(keep = 'first', inplace = False)
```

```python
df['comb'] = df['director_name']+' '+df['genres']+' '+df['actor_1_name']+' '+df['actor_2_name']+' '+df['actor_3_name']+' '+df['keywords']
df
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['comb'].values.astype('U'))
similarity = cosine_similarity(count_matrix)
similarity.shape
```

```python
def title_from_index(index):
    return df[df.index == index]['movie_title'].values[0]
def index_from_title(movie):
    return df[df.movie_title == movie].index.values[0]
```

```python
movie_user_likes = "the avengers"
similar_movies = []
movie_index = index_from_title(movie_user_likes)
similar_movies = similar_movies + list(enumerate(similarity[movie_index]))
sorted_sim_movies = sorted(similar_movies, key = lambda x:x[1], reverse = True)
```

```python
cnt = 0
for movie in sorted_sim_movies:
    if title_from_index(movie[0]) == movie_user_likes:
        continue
    else:
        print(title_from_index(movie[0]))
        cnt = cnt + 1
        if cnt > 20: 
            break
```
