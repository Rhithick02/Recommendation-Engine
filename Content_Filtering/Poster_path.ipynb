```python
import pandas as pd
import numpy as np
from tmdbv3api import Movie, TMDb
import json
import requests
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import time
```

```python
url = "https://image.tmdb.org/t/p/original"
tmdb = TMDb()
tmdb.api_key = 'deb477a3c3471f6e4e39aff4352f0806'
tmdb_movie = Movie()
def scrape(x):
    try:
        result = tmdb_movie.search(x)
        m_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(m_id, tmdb.api_key))
        data_json = response.json()
        response.close()
        return url + str(data_json['poster_path'])
    except:
        return np.NaN
```

```python
df = pd.read_csv('recommend.csv')
df
```

```python
df = df.drop('recommend', axis = 1)
df
```

```python
df['poster_path'] = df['movie_title'].map(lambda x: scrape(x))
df
```
