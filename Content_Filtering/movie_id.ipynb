```python
import pandas as pd
from tmdbv3api import Movie, TMDb
import numpy as np
```

```python
tmdb = TMDb()
tmdb.api_key = 'deb477a3c3471f6e4e39aff4352f0806'
tmdb_movie = Movie()
def scrape(x):
    try:
        result = tmdb_movie.search(x)
        m_id = result[0].id
        return m_id
    except:
        return np.NaN
```

```python
df = pd.read_csv('poster_path.csv')
df
```

```python
df = df.drop('poster_path', axis = 1)
df
```

```python
df['movie_id'] = df['movie_title'].map(lambda x: scrape(x))
df
```
